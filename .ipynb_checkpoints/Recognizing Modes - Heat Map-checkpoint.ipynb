```python
mzi_lower, mzi_upper = mlt.envPeak(data1.mzi.values, delta=0.2, sg_order=1)

plt.plot(data1.mzi,label='MZI')
plt.plot(mzi_lower,'r',label='Lower Envelope')
plt.plot(mzi_upper,'m',label='Upper Envelope')
plt.ylabel('Trans. (V)')
plt.legend(loc='best')

plt.show()
plt.close()
```

```python
peak_hcn = 2 # Peak 8 P-Branch λ0=1548.19

λ0 = 1548.19 # in nm
freq0 = 1e-3*c/λ0 # in THz
print('λ0 = {:.6g} nm, ν0 = {:.6g} THz'.format(λ0, freq0))

idx_central = np.argmin(np.abs(ind_min_hcn[peak_hcn] - ind_peaks))
range_vector = (np.arange(0, len(ind_peaks)) - idx_central)/2 # Divided by 2 because the separation between 2 peaks is equal to FSR/2
```

```python
# SiO2 Sellmeier equation
# -----------------------
B1, B2, B3 = 0.6961663, 0.4079426, 0.8974794
C1, C2, C3 = 68.4043, 116.2414, 9896.161 # in nm

x = symbols('x')
n = sqrt(1+(B1*x**2)/(x**2-C1**2)+(B2*x**2)/(x**2-C2**2)+(B3*x**2)/(x**2-C3**2))
ng = n - x*diff(n, x)

# Path difference MZI
# -------------------
L = 1.49e-3 # in km

print('ΔL_mzi = {0:.3} m'.format(1e3*L))

# Valid values for C-Band (1530-1565nm):
# --------------------------------------
S0 = 0.082 # Zero Dispersion Slope (S0) in ps/(nm².km)
Lamb0 = 1280 # Zero Dispersion Wavelength (λ0) in nm

D = S0/4*(x-Lamb0**4/x**3) # in ps/(nm.km)

β1 = 1e15*(ng/c) # ps/km
D1_mzi = 1/β1/L # in THz

β2 = -1e3*(x**2/2/π/c)*D # in ps²/km
D2_mzi = -(2*π*β2/β1)*(1/β1/L)**2 # in THz

β3 = 1e6*x**3/(2*π*c)**2*(2*D+x*S0) # in ps³/km
D3_mzi = (4*π**2/L**3/β1**5)*(3*β2**2-β1*β3) # in THz
```

```python
D1_mzi = lambdify(x, D1_mzi)(λ0) # FSR for MZI in THz
D2_mzi = lambdify(x, D2_mzi)(λ0) # D2/2pi for MZI in THz
D3_mzi = lambdify(x, D3_mzi)(λ0) # D3/2pi for MZI in THz

print('D1_mzi = {:.4g} MHz, D2_mzi = {:.3g} Hz, D3_mzi = {:.3g} μHz'.format(1e6*D1_mzi, 1e12*D2_mzi, 1e18*D3_mzi))

freq_r = freq0 + D1_mzi*range_vector + D2_mzi/2*range_vector**2 + D3_mzi/6*range_vector**3

freq_ifunc = interpolate.interp1d(data1.index[ind_peaks], freq_r) # y(x) = ...(xv,yv)
data1 = data1.iloc[min(ind_peaks):max(ind_peaks),:].copy()

data1['freq'] = freq_ifunc(data1.index)
data1.reset_index(drop = True, inplace = True)
data1.head()
```

```python
def cav_lorentzian(Delta, delta0, eta, kappa):
    den = 4*(Delta-delta0)**(2)+kappa**(2)
    num = 4*(Delta-delta0)**(2)+(1-2*eta)**(2)*kappa**(2)
    func =  num/den
    return func

def normalize_fit(peak_index):
    ind_0 = ind_min[peak_index] #índice do mínimo, veja na figura anterior
    delta_ind = 15*np.where(data1.cav_norm[ind_0:]>(data1.cav_norm[ind_0]+(np.max(data1.cav_norm)-data1.cav_norm[ind_0])/2))[0][0]
    ind_i = ind_0 - int(delta_ind)
    ind_f = ind_0 + int(delta_ind)
    
    freq_res = 1e6*(data1.freq[int(ind_i):int(ind_f)] - data1.freq[ind_0]) # in MHz
    mzi_res = data1.mzi_smooth[int(ind_i):int(ind_f)]
    cav_res = data1.cav_norm[int(ind_i):int(ind_f)]

    nstd = 0.5 # numero de desvios-padrão para se ignorar
    background_mask = cav_res>(np.max(cav_res)-nstd*np.std(cav_res)) # remove the ressonance peak 
    
    background = np.mean(cav_res[background_mask])
    cav_norm = cav_res/background
    
    guess = [0, 0.1, 10]
    pfit, pcov = optimize.curve_fit(cav_lorentzian, freq_res, cav_norm, p0=guess, maxfev=1500)
    
    return pfit
```

```python
Q_vec = []
ν_vec = []

for index in progressbar.progressbar(np.arange(0, len(ind_min))):
    fit = normalize_fit(peak_index = index)
    eta = fit[1]
    kappa = fit[2]
    kappa_i = kappa*(1-eta)
    kappa_e = kappa*eta
    freq_peak = data1.freq[ind_min[index]]
    Q_t = freq_peak/kappa # Total quality factor in millions
    Q_vec.append(Q_t)
    ν_vec.append(freq_peak)

Q_vec = np.array(Q_vec) # in millions
ν_vec = np.array(ν_vec) # in THz
```
