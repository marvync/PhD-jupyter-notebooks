```python
import os
import glob
import pandas as pd
import seaborn as sns
    
from matplotlib import rcParams
from matplotlib import pyplot as plt
```

```python
sns.set(context='talk', style='ticks', palette='dark', font='sans-serif', font_scale=1.2, color_codes=True, rc=None)
rcParams['svg.fonttype'] = 'none'
```

```python
path = os.getcwd()

flist = sorted(glob.glob(os.path.join(path,'*68*.pkl')), key=os.path.getmtime)
flist
```

```python
for ii in range(len(flist)):
    N=ii
    fname = flist[N]
    print(fname)

    ofc = pd.read_pickle(fname)
    ofc.head()
    ofc.plot(x='λ (nm)', y='Power (dBm)', figsize=(12,4))
    plt.tight_layout()

    plt.savefig(fname[0:-4]+'.svg')
```
