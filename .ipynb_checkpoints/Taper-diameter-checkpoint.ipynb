```python
import numpy as np

from matplotlib import pyplot as plt
from SettingPLOTs import *
from scipy import constants

π = constants.pi
```

```python
d0 = 125e-6 # initial fiber diameter
Lhot = 1e-3*np.arange(2,6) # hot-zone
d = 1e-6*np.arange(0.5, 10, 0.1) # final fiber diameter
```

```python
λ = 1.55 # in μm
n = 1.4682
dsmf = 2.405*λ/π/np.sqrt(n**2-1) # Diameter for single-mode condition
print(f'dsmf = {dsmf:.2f} μm')
```

```python
plt.figure(figsize=cm2inch(10,7))
for L0 in Lhot:
    x = 2*L0*np.log(d0/d) # total stretching distance
    plt.plot(1e6*d, 1e3*x, label='{:.0f} mm'.format(1e3*L0))

# plt.vlines(x=dsmf, ymin=9, ymax=55, lw=1.1, ls='--', color='k') # single-mode condition
plt.ylabel(r'Total stretching distance, $x$ (mm)')
plt.xlabel(r'Taper diameter, $d$ (μm)')
plt.legend(loc='best')

# plt.savefig('Figures/Taper-diameters.svg')
plt.show()
```
