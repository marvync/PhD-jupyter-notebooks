### Riccati-Bessel functions

```python
import numpy as np

from scipy import special as sp
from matplotlib import pyplot as plt
from SettingPLOTs import *
```

```python
l = 4
label = np.arange(l+2)
κr = np.arange(0, 7.1, .01)

Ric_Bes = []
Ric_Neu = []

for ii in κr:
    Ric_Bes.append(sp.riccati_jn(l, ii)[0])
    Ric_Neu.append(sp.riccati_yn(l, ii)[0])


fig, axs = plt.subplots(1,2, figsize=cm2inch(16,5))

axs[0].plot(κr, Ric_Bes)
axs[0].set_xlabel('$κr$')
axs[0].set_ylabel('$ψ_l(κr)$')

axs[1].plot(κr, Ric_Neu)
axs[1].set_xlabel('$κr$')
axs[1].set_ylabel('$χ_l(κr)$')
axs[1].set_ylim(-10,2)
plt.legend(label, loc='center left', bbox_to_anchor=(1, 0.5))

plt.tight_layout()
# plt.savefig('./Figures/Riccati-Bessel-functions.svg')
plt.show()
```
