### Riccati-Bessel functions

```python
import numpy as np

from scipy import special as sp
from matplotlib import pyplot as plt
from SettingPLOTs import *
```

```python
l = 4
label = np.arange(l+2)
ρ = np.arange(0, 7.1, .01)

Ric_Bes = []
Ric_Neu = []

for ii in ρ:
    Ric_Bes.append(sp.riccati_jn(l, ii)[0])
    Ric_Neu.append(-sp.riccati_yn(l, ii)[0])

fig, axs = plt.subplots(1,2, figsize=cm2inch(16,5))

axs[0].plot(ρ, Ric_Bes)
axs[0].set_xlabel('$ρ$')
axs[0].set_ylabel('$ψ_l(ρ)$')

axs[1].plot(ρ, Ric_Neu)
axs[1].set_xlabel('$ρ$')
axs[1].set_ylabel('$χ_l(ρ)$')
axs[1].set_ylim(-2,10)
plt.legend(label, loc='center left', bbox_to_anchor=(1, 0.5))

plt.tight_layout()
# plt.savefig('./Figures/Riccati-Bessel-functions.svg')
plt.show()
```

```python
li, lf = 1, 6
l = range(li,lf)
ρ = np.arange(0.01, 8, .05)

ψ = lambda l, ρ: ρ*sp.spherical_jn(l, ρ)
χ = lambda l, ρ: -ρ*sp.spherical_yn(l, ρ)

fig, axs = plt.subplots(1,2, figsize=cm2inch(16,5))

for ii in l:
    ψ_vec = []
    χ_vec = []
    for jj in ρ:
        ψ_vec.append(ψ(ii, jj))
        χ_vec.append(χ(ii, jj))
    axs[0].plot(ρ, ψ_vec)
    axs[1].plot(ρ, χ_vec)

axs[0].set_xlabel('$ρ$')
axs[0].set_ylabel('$ψ_l(ρ)$')

axs[1].set_xlabel('$ρ$')
axs[1].set_ylabel('$χ_l(ρ)$')
axs[1].set_ylim(-2,10)
plt.legend(l, loc='center left', bbox_to_anchor=(1, 0.5))

plt.tight_layout()
# plt.savefig('./Figures/Riccati-Bessel-functions.svg')
plt.show()
```
